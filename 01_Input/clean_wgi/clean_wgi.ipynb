```python
import pandas as pd
```

```python
# Read the initial Excel file
ancor_pa = pd.read_excel('C:/Users/Ardhi/OneDrive - SIPA, Columbia University/MPA SIPA/Work/UNDP/Anticorruption and Just Energy/Energy-Corruption-main/Clean/wgidataset.xlsx',sheet_name="VoiceandAccountability")
ancor_ps = pd.read_excel('C:/Users/Ardhi/OneDrive - SIPA, Columbia University/MPA SIPA/Work/UNDP/Anticorruption and Just Energy/Energy-Corruption-main/Clean/wgidataset.xlsx',sheet_name="Political StabilityNoViolence")
ancor_ge = pd.read_excel('C:/Users/Ardhi/OneDrive - SIPA, Columbia University/MPA SIPA/Work/UNDP/Anticorruption and Just Energy/Energy-Corruption-main/Clean/wgidataset.xlsx',sheet_name="GovernmentEffectiveness")
ancor_rq = pd.read_excel('C:/Users/Ardhi/OneDrive - SIPA, Columbia University/MPA SIPA/Work/UNDP/Anticorruption and Just Energy/Energy-Corruption-main/Clean/wgidataset.xlsx',sheet_name="RegulatoryQuality")
ancor_rl = pd.read_excel('C:/Users/Ardhi/OneDrive - SIPA, Columbia University/MPA SIPA/Work/UNDP/Anticorruption and Just Energy/Energy-Corruption-main/Clean/wgidataset.xlsx',sheet_name="RuleofLaw")
ancor_cc = pd.read_excel('C:/Users/Ardhi/OneDrive - SIPA, Columbia University/MPA SIPA/Work/UNDP/Anticorruption and Just Energy/Energy-Corruption-main/Clean/wgidataset.xlsx',sheet_name="ControlofCorruption")


ancor_cc.head()
```

```python
# Rename columns for consistency and ease of use
ancor_pa = ancor_pa.rename(columns={
    'Country/Territory': 'Country',
    'Code': 'Country Code'
})
ancor_ps = ancor_ps.rename(columns={
    'Country/Territory': 'Country',
    'Code': 'Country Code'
})
ancor_ge = ancor_ge.rename(columns={
    'Country/Territory': 'Country',
    'Code': 'Country Code'
})
ancor_rq = ancor_rq.rename(columns={
    'Country/Territory': 'Country',
    'Code': 'Country Code'
})
ancor_rl = ancor_rl.rename(columns={
    'Country/Territory': 'Country',
    'Code': 'Country Code'
})
ancor_cc = ancor_cc.rename(columns={
    'Country/Territory': 'Country',
    'Code': 'Country Code'
})

ancor_cc.head()
```

```python
# Extract necessary columns for the range of years
id_columns = ['Country', 'Country Code']
year_columns_pa = [col for col in ancor_pa.columns if col.endswith('_Estimate')]
year_columns_ps = [col for col in ancor_ps.columns if col.endswith('_Estimate')]
year_columns_ge = [col for col in ancor_ge.columns if col.endswith('_Estimate')]
year_columns_rq = [col for col in ancor_rq.columns if col.endswith('_Estimate')]
year_columns_rl = [col for col in ancor_rl.columns if col.endswith('_Estimate')]
year_columns_cc = [col for col in ancor_cc.columns if col.endswith('_Estimate')]


ancor_pa_new = ancor_pa[id_columns + year_columns_pa]
ancor_ps_new = ancor_ps[id_columns + year_columns_ps]
ancor_ge_new = ancor_ge[id_columns + year_columns_ge]
ancor_rq_new = ancor_rq[id_columns + year_columns_rq]
ancor_rl_new = ancor_rl[id_columns + year_columns_rl]
ancor_cc_new = ancor_cc[id_columns + year_columns_cc]
ancor_pa_new.head()
```

```python
# Melt the DataFrame to transform the data structure
ancor_pa_new = ancor_pa_new.melt(id_vars=id_columns, var_name='Year', value_name='Value')
ancor_ps_new = ancor_ps_new.melt(id_vars=id_columns, var_name='Year', value_name='Value')
ancor_ge_new = ancor_ge_new.melt(id_vars=id_columns, var_name='Year', value_name='Value')
ancor_rq_new = ancor_rq_new.melt(id_vars=id_columns, var_name='Year', value_name='Value')
ancor_rl_new = ancor_rl_new.melt(id_vars=id_columns, var_name='Year', value_name='Value')
ancor_cc_new = ancor_cc_new.melt(id_vars=id_columns, var_name='Year', value_name='Value')

ancor_pa_new[['Year', 'Parameter']] = ancor_pa_new['Year'].str.split('_', expand=True)
ancor_ps_new[['Year', 'Parameter']] = ancor_ps_new['Year'].str.split('_', expand=True)
ancor_ge_new[['Year', 'Parameter']] = ancor_ge_new['Year'].str.split('_', expand=True)
ancor_rq_new[['Year', 'Parameter']] = ancor_rq_new['Year'].str.split('_', expand=True)
ancor_rl_new[['Year', 'Parameter']] = ancor_rl_new['Year'].str.split('_', expand=True)
ancor_cc_new[['Year', 'Parameter']] = ancor_cc_new['Year'].str.split('_', expand=True)
```

```python
# Reorder columns
ancor_pa = ancor_pa_new[['Country', 'Country Code', 'Year', 'Value', 'Parameter']]
ancor_ps = ancor_ps_new[['Country', 'Country Code', 'Year', 'Value', 'Parameter']]
ancor_ge = ancor_ge_new[['Country', 'Country Code', 'Year', 'Value', 'Parameter']]
ancor_rl = ancor_rl_new[['Country', 'Country Code', 'Year', 'Value', 'Parameter']]
ancor_rq = ancor_rq_new[['Country', 'Country Code', 'Year', 'Value', 'Parameter']]
ancor_cc = ancor_cc_new[['Country', 'Country Code', 'Year', 'Value', 'Parameter']]

ancor_pa['Indicator Code'] = 'ancor-pa'
ancor_ps['Indicator Code'] = 'ancor-ps'
ancor_ge['Indicator Code'] = 'ancor-ge'
ancor_rl['Indicator Code'] = 'ancor-rl'
ancor_rq['Indicator Code'] = 'ancor-rq'
ancor_cc['Indicator Code'] = 'ancor-cc'
```

```python
# Save the transformed DataFrame to a new Excel file
output_file_path1 = 'ancor_pa.xlsx'
output_file_path2 = 'ancor_ps.xlsx'
output_file_path3 = 'ancor_ge.xlsx'
output_file_path4 = 'ancor_rl.xlsx'
output_file_path5 = 'ancor_rq.xlsx'
output_file_path6 = 'ancor_cc.xlsx'
ancor_pa.to_excel(output_file_path1, index=False)
ancor_ps.to_excel(output_file_path2, index=False)
ancor_ge.to_excel(output_file_path3, index=False)
ancor_rl.to_excel(output_file_path4, index=False)
ancor_rq.to_excel(output_file_path5, index=False)
ancor_cc.to_excel(output_file_path6, index=False)
```
